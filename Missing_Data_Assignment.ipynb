```python
df=df.dropna(subset=["ManagerID"])
p =df.isna().mean() * 100
p.round()
```

```python
w = df["Position"].unique()
s = {}

for i in w:
    salary = df.loc[df['Position'] == i, 'Salary'].mean()
    s[i] = salary

for row in df.iterrows():
    position = row[1]['Position']
    salary = s[position]
    df.loc[row[0], 'Salary'] = salary   
p =df.isna().mean() * 100
p.round()
```

```python
df=df.dropna(subset=["Salary"])
p =df.isna().mean() * 100
p.round()
```

```python
import datetime
w = df["Position"].unique()
a = df.dropna(subset=["DateofTermination"])

a["DateofHire"] = pd.to_datetime(a["DateofHire"], format='%m/%d/%Y')
a["DateofTermination"] = pd.to_datetime(a["DateofTermination"], format='%m/%d/%Y', errors='coerce')
a["Tenure"] = (a["DateofTermination"].dt.year - a["DateofHire"].dt.year)
q =a[['Position','Tenure']]
average_tenure_by_position = a.groupby('Position')['Tenure'].mean()
average_tenure_dict = average_tenure_by_position.to_dict()
average_tenure_dict
```

```python
df["DateofHire"] = pd.to_datetime(df["DateofHire"], format='%m/%d/%Y')

for i in df["Position"]:
    if i not in average_tenure_by_position.keys():
        continue
    df['DateofTermination']=df['DateofHire'].dt.year+average_tenure_by_position[i]
    
p =df.isna().mean() * 100
p.round()
```

```python
means = df["Absences"].mean().round()
df['Absences'].fillna(means, inplace=True)
p =df.isna().mean() * 100
p.round()
```

```python

df['Employee_Name'].fillna("NO_NAME", inplace=True)
p =df.isna().mean() * 100
p.round()
```

```python
df['Sex'].fillna("NO_Sex", inplace=True)
p =df.isna().mean() * 100
p.round()
```

```python
df['DOB'].fillna("NO_DOB", inplace=True)
p =df.isna().mean() * 100
p.round()
```
